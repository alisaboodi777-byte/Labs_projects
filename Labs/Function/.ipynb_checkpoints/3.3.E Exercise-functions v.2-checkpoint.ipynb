#### Q1. 
Write a function that takes two numbers, `a` and `b`, converts them to `int`, and then returns the result of multiplying `a` by the negative of `b`.

\

##### Q5.D
Write a function that accepts one parameter, a dictionary. \
The function should print the message "Introducing ______ ______, ________!" \
\
Fill in the blanks with the "first_name", "last_name", and "title" properties of the dictionary

```python
# your code here
def hype_function (d):
    name = d['first_name']
    lastname = d['last_name']
    title = d['title']
    print(f"Introducing {name} {lastname} {title}")

```

```python
d={"first_name":"Mohammed","last_name":"Abdullah","title":"Data Scientist"}

hype_function(d)
```

#### Q6. Problem-Solving Functions

```python
# A. Write a function that finds the longest string in a given list of strings

# your code here
def longest_string_finder(h):
    long=''
    for s in h:
        if len(s) > len(long):
            long = s
    return long

```

```python
s=["dfaf","adfaffaf","grrtsrfsgfgfsgsgf","adfaaa"]

longest_string_finder(s)
```

```python
# B. Write a function to determine whether a given list contains a given value

# your code here
def search_list(a,v):
    if v in a:
        print('True')
    else:
        print("False")

```

```python
ls=[34,90,120,432,647,875]

search_list(ls,432)
```

```python
ls=[34,90,120,432,647,875]

search_list(ls,122)
```

```python
# D. Write a function to count how many times a given string contains a given letter

# your code here
def char_count(s,l):
    count_l = s.count(l)
    return count_l
```
